## Area Plots

```python
import pandas as pd
import numpy as np

%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

pd.options.mode.use_inf_as_na = False
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df_canada = pd.read_excel('Canada.xlsx', sheet_name = 'Canada by Citizenship', skiprows = range(20), skipfooter = 2)
print("Data downloaded and read into dataframe")
```

```python
df_canada.head(5)
```

```python
df_canada.shape
```

```python
df_canada.columns
```

```python
#Dropping the unwanted columns from df

df_canada.drop(['AREA','Coverage', 'Type', 'DEV', 'REG'],axis=1, inplace=True)
df_canada.head(2)
```

```python
#Replacing the columns-name with the names suited for reporting or visulaization
df_canada.rename(columns={'OdName':'Country', 'AreaName': 'Continent', 'RegName':'Region'},inplace = True) # inplace = true , changes the value within the data frame
df_canada.columns
```

```python
df_canada.head(5)
```

```python
#Validate the values whether all the values are string in the dataframe
all(isinstance(column, str) for column in df_canada.columns)
```

```python
#another way to find the datatypes
df_canada.dtypes
```

```python
#Convert the columns to string which are not string data types
df_canada.columns = list(map(str,df_canada.columns))
df_canada.columns
```

```python
#Validating the whether the data types are string or not 
all(isinstance(column, str) for column in df_canada.columns)
```

```python
#Setting the index to column Country
df_canada.set_index('Country', inplace = True)
df_canada.head(5)
```

```python
#Added the columns as Totals

df_canada['Total'] = df_canada.sum(axis=1)

df_canada.head(5)
```

```python
#Number of elements in each dimensions. get the details of columns and rows in the data frame 
df_canada.shape
```

```python
#converting the years columns as strings
years = list(map(str, range (1980, 2014)))
years
```

```python
#Most frequent one is GGplot in data visulaizations

mpl.style.use('ggplot')
```

```python
##Sorting the values based on Total column
df_canada.sort_values(by = 'Total', ascending = False, axis = 0 , inplace = True)
#passing the top 5 values of the dataframe into another df
df_top5 = df_canada.head(5)
#print the top 5 data 
df_top5
```

```python
#Transpose the index as columns and columns as rows
df_top5 = df_top5[years].T
df_top5.head()
```

```python
# area plots are nothing but a line plots only the difference is plots are shaded

df_top5.index = df_top5.index.map(int)
df_top5.plot(kind = 'area', stacked = False, figsize = (15, 8))

plt.title('Top 5 countries - Immigration Trends to Canada')
plt.ylabel('Number of Immigrannts')
plt.xlabel('Years')

plt.show()
```

```python
# area plots are nothing but a line plots which is shaded bit 

# stacked  = True 
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind = 'area', stacked = True, figsize = (15, 8))

plt.title('Top 5 countries - Immigration Trends to Canada')
plt.ylabel('Number of Immigrannts')
plt.xlabel('Years')

plt.show()
```

#### The unstacked area plots provides better visuals than the stacked area plots 

#### There are two types of plotting methods :
 #### - Plotting using the scripting layer (procedural method) - using matplotlib.pyplot as plt
 #### - One is plotting using the Artist layer (Object oriented method) - using Axes instance from Matplotlib (preferred)

## Scripting Layer

```python
df_top5.plot(kind = 'area', alpha= 0.5, figsize = (15, 8))

plt.title('Top 5 countries - Immigration Trends to Canada')
plt.ylabel('Number of Immigrannts')
plt.xlabel('Years')

plt.show()

#alpha - color variation i.e., brightness or lightness of the color
```

### Artist Layer

```python
ax = df_top5.plot(kind = 'area', alpha= 0.33, figsize = (15, 8))

ax.set_title('Top 5 countries - Immigration Trends to Canada')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')

#alpha - color variation brightness or lightness of the color
```

# Bar Plots 

#### Bar plot is a way of representing data where the length of the bars respresents the magnitude/size of the feature/variable. 

#### Bar graphs usually represent the number and categorical variables grouped in intervals.

##### To create bar plot, we pass the one of the two arguments via 'kind' parameter in plot() function
* <b>'kind = bar' </b> creates a vertical bar plot
* <b> 'kind = barh' </b> created a horizontal bar plot

#### Use case : Let's analyze the effects of Iceland's financial crisis between 2008-2011 which is a major political and economic event.

```python
df_iceland = df_canada.loc['Iceland', years]
df_iceland.head(5)
```

```python
# step 2 : bar plotting
df_iceland.plot(kind = 'bar', figsize = (10,6))

plt.xlabel('Year') # add xlabel to the plot
plt.ylabel('Number of Immigrants') # add ylabel to the plot
plt.title('Iceland immigrants to Canada from 1980 to 2013')

plt.show()
```

##### The above bar plot is representing the increase in numbers after 2008 financial crisis in Iceland.
##### We can highlight the gradual increase due to financial crisis with annotate method of the scripting layer. 
##### This method is part of pyplot interface/module in matplotlib

```python
# step 2 : plot data 
df_iceland.plot(kind = 'bar', figsize = (10,6), rot = 90)

plt.xlabel('Year') # add xlabel to the plot
plt.ylabel('Number of Immigrants') # add ylabel to the plot
plt.title('Iceland immigrants to Canada from 1980 to 2013') # add title to the plot 

plt.show()
```

```python
# step 2: plot data
df_iceland.plot(kind='bar', figsize=(10, 6),rot=90)


plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of immigrants') # add y-label to the plot
plt.title('Iceland immigrants to Canada from 1980 to 2013') # add title to the plot

plt.annotate('', 
            xy = (33,71),        #Place where Arrow head is annotated (year 2013, y-value is 71)
            xytext = (28, 20),     # Place where base of arrow is annotated (year 2008, y-valueis 20)
            xycoords = 'data',
            arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3', color = 'darkgreen', lw = 5)
            )


plt.show()
```

```python
# step 2: plot data
df_iceland.plot(kind='bar', figsize=(10, 6),rot=90)


plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of immigrants') # add y-label to the plot
plt.title('Icelandic immigrants to Canada from 1980 to 2013') # add title to the plot

## Annotating the arrow
plt.annotate('', 
            xy = (33,71),        #Place where Arrow head is annotated (year 2013, y-value is 71)
            xytext = (28, 20),     # Place where base of arrow is annotated (year 2008, y-valueis 20)
            xycoords = 'data',
            arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3', color = 'darkgreen', lw = 5)
            )

## Annotating the Text

plt.annotate('2008-2011 Iceland Crisis',
            xy=(28,30),                 # Place where Text annotation will be started (year 2008, y-value is 30)
            rotation = 73,              # based on trail and arrow, we will match the rotation angle to be in parallel with arrow
             va='bottom',               # Horizontal Text Alignment
             ha='left'                  # Vertical Text Alignment
            )

plt.show()
```

#### When we want to present something more practically and the data should be projected horizontally to label the feature on Y-axis and their magnitude on the x-axis.

```python
#Plotting horizontal bars by giving kind as barh
df_iceland.plot(kind = 'barh', figsize = (10,6))

plt.xlabel('Year') # add xlabel to the plot
plt.ylabel('Number of Immigrants') # add ylabel to the plot
plt.title('Iceland immigrants to Canada from 1980 to 2013')

plt.show()
```

#### Use case : Let's get the Horizontal Barchart projected for the top 15 countries  and label each country with the total immigration count. We also want to add labels on horizontal bars with annotation



```python
df_canada.sort_values(by='Total', ascending=True,inplace=True)
df_top15 = df_canada['Total'].tail(15)
df_top15
```

```python
#Plotting horizontal bars by giving kind as barh
df_top15.plot(kind = 'barh', figsize = (10,12))

plt.xlabel('Year') # add xlabel to the plot
plt.ylabel('Number of Immigrants') # add ylabel to the plot
plt.title('Iceland immigrants to Canada from 1980 to 2013')

plt.show()
```

```python
#Plotting horizontal bars by giving kind as barh
df_top15.plot(kind = 'barh', figsize = (10,12))

plt.xlabel('Number of Immigrants') # add xlabel to the plot
plt.ylabel('Countries') # add ylabel to the plot
plt.title('Top 15 countries immigration to canada')

for index, value in enumerate(df_top15):
    label = format(int(value), ',')
    
    plt.annotate(label, xy=(value-60000, index -0.10), rotation = 0, color='white')

plt.show()
```
